таблицы в отсортированном виде. Параметрами функции являются список, функция,
которая определяет порядок сортировки (значение параметра key в sort), и параметр,
определяющий, выполняется сортировка по возрастанию или убыванию. Используя эту функцию, получите словарь, ключами которого являются фамилии, а значениями сумма баллов. 

```python
import pandas as pd
def visual(lst, usr_key, revers=False):
    list(map(lambda l: l[1:],lst))
    res_lst = sorted(lst, key=usr_key, reverse=revers)
    df = pd.DataFrame(res_lst)
    dct = {i[0]: sum(i[1:]) for i in res_lst}
    print(dct)
    print(df)
usr_lst = [['Иванов', 3,5,4,4,3], ['Петров',4,5,4,4,5], ['Сидоров',3,4,3,3,3], ['Иванов', 4,5,4,4,4], ['Николаев',5,4,4,4,3]]
visual(usr_lst, lambda lst: lst[1])
```

Написать скрипт, который из строк текстового файла делает столбцы CSV файла. 

Скрипт принимает через командную строку 2 параметра: 

1) первый параметр - имя файла из которого берется текст (файл имеет кодировку UTF8; нет переносов слов на другую строку). В текстовом файле содержатся строки вида:

Имя столбца: Значение1, Значение из нескольких слов2, ЗначениеN


2) второй параметр - имя файла, в который сохраняется файл CSV (с разделителем ; и с заголовком, файл имеет кодировку UTF8).

Пример исходного файла:
Фамилия:Иванов,Петров,,Сидоров
Регион:Центр,Северный Кавказ,Дальний Восток,Север
Оплата:100,90,150,70

Результат в CSV:
Фамилия;Регион;Оплата
Иванов;Центр;100
Петров;Северный Кавказ;90
;Дальний Восток;150
Сидоров;Север;70

Скрипт должне иметь вид tocsv_Ivanov.py (вместо Ivanov - ваша фамилия).  Функции, связанные с обработкой файлов и заменой в тексте, должны находится в отдельном модуле modul_Ivanov.py (вместо Ivanov - ваша фамилия).

```python
import csv
def f(filename_txt, filename_csv):
    data_lst = []
    f = open(filename_txt, 'r', encoding='UTF-8').read()
    data_lst = f.split()
    with open('file1.csv', 'w', encoding='UTF-8') as f:
        write = csv.writer(f)
        write.writerows([i.split(';') for i in data_lst])
    with open('file1.csv', 'r', encoding='UTF-8') as f:
        reade = csv.reader(f, delimiter=';')
        for i in reade:
            print(i)

f('file.txt', 'csv_file')
```

В 2.2 

Реализовать функцию repl , которая принимает на вход строку и набор заранее неизвестных параметров. Результатом функции является строка, в которой слова совпадающие с именами параметров заменены на значения параметров.

Пример: repl('replace my val abc', my='s1', abc='fff') -> 'replace s1 val fff'

```python
def repl(s, **params):
    s = s.split()
    res = ''
    for i in range(len(s)):
        if s[i] in params:
            s[i] = params[s[i]]
    for i in s:
        res += i + ' '
    return res
print(repl('replace my val abc', my='s1', abc='fff'))
```
